### Sync Git repo with Docker

```python
cd C:\SHIP\DATAFILES
git clone https://github.com/lauramarion/data_pipeline
cd data_pipeline
```

### Adapt dagster files to new structure
Move dockerfile and requierements.txt to the root folder (data_pipeline).  
Update ENTRYPOINT in dockerfile with new url structure :

```python
# 3. Execution: Point to the files inside the 'dagster/' subfolder
# The Code Server will run the definitions file found at /app/dagster/definitions.py
ENTRYPOINT ["dagster", "dev", "-f", "/app/dagster/definitions.py", "--host", "0.0.0.0"]
```

### Get Dagster up and running
Rebuild dagster image

```python
docker rm dagster
docker build -t dagster_pipeline .
```

then run the container

```python
# bash
docker run -d \
  --name dagster \
  --network metabase-net \
  -e NOCODB_DB_PASSWORD="299centblues" \
  -p 3002:3000 \
  dagster_pipeline \
```
